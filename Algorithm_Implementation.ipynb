```python
import time

 # Naive String Matching Algorithm
def naive_string_matching(text, pattern):

# Store the matching positions
    n = len(text)
    m = len(pattern)
    results = []
   # Start measuring execution time
    start_time = time.perf_counter()
 # Slide the pattern over the text one position at a time
    for i in range(n - m + 1):
        j = 0
# Compare the pattern with the current text window
        while j < m and text[i + j] == pattern[j]:
            j = j + 1

        # If j reaches the length of the pattern, a match is found
        if j == m:
            results.append(i)
# If mismatch occurs, the algorithm automatically
 #shifts the pattern by one position and continues
    # End the timing mechanism
    end_time = time.perf_counter()

# Calculate total execution time
    execution_time = end_time - start_time
    return results, execution_time

# STEP: Testing the code with a simple example

sample_text = "A" * 1_000_000 + "B"
sample_pattern = "AAAAB"

indices, duration = naive_string_matching(sample_text, sample_pattern)

print("-- Simple Test Case --")
print(f"Text length: {len(sample_text)} characters")
print(f"Text preview: {sample_text[:50]}...")
print(f"Pattern: {sample_pattern}")
print(f"Pattern found at indices: {indices}")
print(f"Execution Time: {duration:.8f} seconds")
```

```python
import time

# Build the LPS (Longest Prefix Suffix) array
# to avoid repeated comparisons
def compute_lps(pattern):
 m = len(pattern)
 # Initialize the LPS array
 lps = [0]*m
 # Length of the previous longest prefix suffix
 length = 0
 i = 1
 # Construct the LPS array
 while i < m:
     # If characters match, extend the prefix length
     if pattern[i] == pattern[length]:
         length += 1
         lps[i] = length
         i += 1
     else:
         # Handle mismatch using previous LPS values
         if length != 0:
             length = lps[length-1]
         else:
             # No prefix suffix match found
             lps[i] = 0
             i += 1
 return lps
def kmp_search(text, pattern):
 # Start measuring execution time
 start_time = time.perf_counter()
 n = len(text)
 m = len(pattern)
 # Build the LPS array
 lps = compute_lps(pattern)
 # Store all matching positions
 results = []
 i = 0
 j = 0
 # Traverse the text
 while i < n:
     # Compare characters between text and pattern
     if text[i] == pattern[j]:
         i += 1
         j += 1
     # Full pattern match found
     if j == m:
         results.append(i - j)
         # Use the LPS array to shift efficiently
         j = lps[j - 1]
     # Handle mismatch after partial matching
     elif i < n and text[i] != pattern[j]:
         # Use LPS array to avoid unnecessary comparisons
         if j != 0:
             j = lps[j - 1]
         else:
             # Move to the next character in the text
            i += 1
 # End measuring execution time
 end_time = time.perf_counter()

 # Calculate execution time
 duration = end_time - start_time
 return results, duration

sample_text = "A" * 1_000_000 + "B"
sample_pattern = "AAAAB"

indices, duration = kmp_search(sample_text, sample_pattern)

print("-- Simple Test Case --")
print(f"Text length: {len(sample_text)} characters")
print(f"Text preview: {sample_text[:50]}...")
print(f"Pattern: {sample_pattern}")
print(f"Pattern found at indices: {indices}")
print(f"Execution Time: {duration:.8f} seconds")
```

```python
import time

# Generate a large text and pattern for worst-case performance testing
sample_text = "A" * 1_000_000 + "B"
sample_pattern = "AAAAB"
# Measure execution time for the Naive algorithm
def measure_naive(text, pattern):
    start = time.perf_counter()

    # Execute the Naive string matching algorithm
    result = naive_string_matching(text, pattern)
    end = time.perf_counter()
    # Extract matches if the function returns additional values
    if isinstance(result, tuple):
        matches = result[0]
    else:
        matches = result
    # Return matches and execution time in milliseconds
    return matches, (end - start) * 1000
# Measure execution time for the KMP algorithm
def measure_kmp(text, pattern):
    start = time.perf_counter()
    # Execute the KMP search algorithm
    matches = kmp_search(text, pattern)
    end = time.perf_counter()
    return matches, (end - start) * 1000

# Repeat the experiment 5 times for more accurate results
runs = 5
naive_times = []
kmp_times = []
naive_matches = []
kmp_matches = []

# Run both algorithms using the same input
for _ in range(runs):
    naive_matches, t1 = measure_naive(sample_text, sample_pattern)
    kmp_matches, t2 = measure_kmp(sample_text, sample_pattern)
    # Store execution times
    naive_times.append(t1)
    kmp_times.append(t2)

# Calculate average execution times
avg_naive = sum(naive_times) / runs
avg_kmp = sum(kmp_times) / runs
# Display performance evaluation results
print("\n=== Time Measurement Results (Worst Case) ===")
print(f"Text length: {len(sample_text)}")
print(f"Pattern: {sample_pattern}\n")

print("Naive Algorithm:")
print(f"Matches: {len(naive_matches)}")
print(f"Times (ms): {[f'{t:.4f}' for t in naive_times]}")
print(f"Average: {avg_naive:.4f} ms\n")

print("KMP Algorithm:")
print(f"Matches: {len(kmp_matches)}")
print(f"Times (ms): {[f'{t:.4f}' for t in kmp_times]}")
print(f"Average: {avg_kmp:.4f} ms\n")

# Confirm that both algorithms used identical inputs
print("Both algorithms used the same text and pattern.")
```

```python
def measure_algorithm(algorithm_func, text, pattern, runs=5):
    times = []
    result = None

    for _ in range(runs):
        start = time.time()
        result = algorithm_func(text, pattern)
        end = time.time()
        times.append((end - start) * 1000)   # convert to milliseconds

    average_time = sum(times) / len(times)
    return result, times, average_time


# Print results in a clean format
def print_case_results(case_name, text, pattern):
    print(f"\n=== Time Measurement Results ({case_name}) ===")
    print(f"Text length: {len(text)}")
    print(f"Pattern: {pattern}\n")

    # Naive
    naive_result, naive_times, naive_avg = measure_algorithm(
        naive_string_matching, text, pattern
    )

    print("Naive Algorithm:")
    if isinstance(naive_result, list):
        print(f"Matches: {len(naive_result)}")
        print(f"Indices: {naive_result[:10]}")   # only show first 10 if many
    else:
        print(f"Result: {naive_result}")
    print(f"Times (ms): {[f'{t:.4f}' for t in naive_times]}")
    print(f"Average: {naive_avg:.4f} ms\n")

    # KMP
    kmp_result, kmp_times, kmp_avg = measure_algorithm(
        kmp_search, text, pattern
    )

    print("KMP Algorithm:")
    if isinstance(kmp_result, list):
        print(f"Matches: {len(kmp_result)}")
        print(f"Indices: {kmp_result[:10]}")
    else:
        print(f"Result: {kmp_result}")
    print(f"Times (ms): {[f'{t:.4f}' for t in kmp_times]}")
    print(f"Average: {kmp_avg:.4f} ms\n")

    print("Both algorithms used the same text and pattern.")
    print("-" * 50)

# ===== Define the 3 cases =====
# Best case: pattern found at the beginning
best_text = "AAAAB" + "A" * 100000
best_pattern = "AAAAB"

# Average case: pattern found in the middle
average_text = "A" * 500000 + "AAAAB" + "A" * 500000
average_pattern = "AAAAB"

# Worst case: many comparisons before match at the end
worst_text = "A" * 1000000 + "B"
worst_pattern = "AAAAB"


# Shakespeare inputs

shakespeare_text_1 = full_text[:50000]
shakespeare_text_2 = full_text[:100000]
shakespeare_pattern = "king"


# all cases
print_case_results("Best Case", best_text, best_pattern)
print_case_results("Average Case", average_text, average_pattern)
print_case_results("Worst Case", worst_text, worst_pattern)

print_case_results("Shakespeare Input 50K", shakespeare_text_1, shakespeare_pattern)
print_case_results("Shakespeare Input 100K", shakespeare_text_2, shakespeare_pattern)
```
